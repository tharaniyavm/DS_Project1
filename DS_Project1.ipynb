**1. Source data loaded from CSV**


```python
import pandas as pd
#Data frame _ 2D
#Pulling csv (comma separated values) file
restaurantdf = pd.read_csv(r"C:\Users\Tharaniya\Documents\GUVI\DS_Project1\Uber_Eats_data.csv",encoding='utf-8')
display(restaurantdf.head(10)) 
```

```python
import pandas as pd
#Data frame _ 2D
#Pulling json  file
ordersdf = pd.read_json(r"C:\Users\Tharaniya\Documents\GUVI\DS_Project1\orders.json")
display(ordersdf.head(10)) 
```

# Data cleaning

```python
# Replaces the corrupted string with 'Cafe'
# Removes everything except letters (A-Z), numbers (0-9), and spaces
restaurantdf['name'] = restaurantdf['name'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

display(restaurantdf.head(10)) 
```

```python
# rename the existing  columns to meaning full names

restaurantdf.rename(columns={'name': 'restaurant_name'}, inplace=True)
restaurantdf.rename(columns={'approx_cost(for two people)': 'approx_two_person_cost'}, inplace=True)
restaurantdf.rename(columns={'listed_in(type)': 'category'}, inplace=True)
restaurantdf.rename(columns={'listed_in(city)': 'locality'}, inplace=True)
display(restaurantdf.head())
```

```python
# Finding the Duplicates sum
print(len(restaurantdf))
restaurantdf.duplicated().sum()
```

```python
# Finding the Duplicate rows
print(f"Number of duplicate sets (where at least one duplicate exists): {restaurantdf.duplicated(keep=False).sum()}")
display(restaurantdf[restaurantdf.duplicated(keep=False)].sort_values(['restaurant_name', 'locality']))
```

```python
# Drop Duplicates
print(restaurantdf.duplicated().sum())
restaurantdf.drop_duplicates(inplace=True)
print(restaurantdf.duplicated().sum())
```

```python
# Identifies the NULL values
restaurantdf.info() # There are no missing values or NULL values in the data set
```

```python
# Splits  data "4.5/5" into ["4.5", "5"] and keeps "4.5"
restaurantdf['rate'] = restaurantdf['rate'].str.split('/').str[0]
restaurantdf
```

```python
# Find the Unique Rates
restaurantdf['rate'].unique()
```

```python
# Removes the spaces in the rate data
restaurantdf['rate'] = restaurantdf['rate'].str.strip()
```

```python
# Replacing new with NULL values
restaurantdf['rate'] = restaurantdf['rate'].replace('NEW', pd.NA)
restaurantdf['rate'] = pd.to_numeric(restaurantdf['rate'], errors='coerce')
display(restaurantdf.head())
```

```python
# Converting rata data type into Float
restaurantdf['rate'] = restaurantdf['rate'].astype(float)
```

```python
# Identifing the NULL values in the rows
restaurantdf.info()
```

```python
# Removing comma in the cost data
restaurantdf["approx_two_person_cost"]=restaurantdf["approx_two_person_cost"].str.replace("," , "")
restaurantdf
```

```python
# converting cost data type to int
restaurantdf["approx_two_person_cost"] = restaurantdf['approx_two_person_cost'].astype(int)
restaurantdf.info()
```

```python
# Spliting phone number into primary and secondary phone
a=restaurantdf["phone"].str.split("\n",expand=True) #converts
display(a)
```

```python
#converts column datatype to string

restaurantdf["primary_phone"]=a[0].astype(str)
restaurantdf["secondary_phone"]=a[1].astype(str)
display(restaurantdf)
```

```python
# Drop the phone column
restaurantdf.drop("phone",axis=1,inplace=True)#drop column permanently
display(restaurantdf)
```

```python
# Specify the new order
new_order = ['restaurant_name', 'primary_phone', 'secondary_phone','location', 'online_order', 'book_table', 'rate', 'votes', 'rest_type', 'dish_liked', 'cuisines', 'approx_two_person_cost', 'category', 'locality']
restaurantdf = restaurantdf[new_order]
display(restaurantdf)
```

```python
# removing \r from the data
restaurantdf["primary_phone"]=restaurantdf["primary_phone"].str.replace("\r" , "")
restaurantdf
```

```python
# Creates a column 'id' from 1 to the total number of rows
restaurantdf.insert(0, 'restaurant_id', range(1, 1 + len(restaurantdf)))
restaurantdf
```

```python
# Removing unvanted characters
ordersdf['restaurant_name'] = ordersdf['restaurant_name'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)

display(ordersdf.head(10)) 
```

# Load the data into sqlite

```python
import sqlite3
import pandas as pd

# 1. Create a connection to the database (it creates the file if it doesn't exist)
conn = sqlite3.connect('ubereats_data.db')

# 2. Write the DataFrame to a table
restaurantdf.to_sql('restaurants', conn, if_exists='replace', index=False)

ordersdf.to_sql('orders', conn, if_exists='replace', index=False)

# 3. Close the connection
conn.close()
```
